# Exersice 6  Strings and Text

A string is usually a bit of text you want to display to someone or "export" out of the program you are writing.

In Python it is not necessary to write or declare variables before you can store anything in them, this is not necessary when working with character strings in Python.

<strong>Concatenation</strong> means joining character strings, this is possible by means of a sum (+), the blanks are marked by placing quotation marks, as shown in the following example

```python
msg1 = 'Hola' + ' ' + 'mundo'
print(msg1)
```

In the last example the string "msg1" contains "Hola mundo"

<strong>Multiplication</strong> this serves to make several copies of a string and is used with the multiplication symbol (*), as shown in the following example

```python
msg2= 'Hola ' * 3
msg3= 'Mundo'
print(msg2+msg3)
```

This is how its work the strings in Python , there are more commands relationated with the strings like <i>len , find , lower , replace ,etc. </i>
